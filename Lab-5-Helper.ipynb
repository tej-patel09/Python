<h1>Use Dynamic List</h1>

```python
# creating dynamic lists Method 1
lst = []
for i in range(3):
    lst.append(int(input("Enter element : ")))
print(lst)
```

```python
# creating dynamic lists Method 2
lst = []
print("Enter -1 to end input")
while True:
    temp = int(input("Enter element : "))
    if temp == -1 :
        break
    else:
        lst.append(temp)
print(lst)
```
